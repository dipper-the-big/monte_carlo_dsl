```python
%matplotlib inline

import plotter
import matplotlib.pyplot as plt
import prettyplotlib as ppl
import numpy as np
plt.rcParams["figure.figsize"] = [10,10]
```

```python
texts = ["steps", "time", "iJumps", "vJumps", "nInterstitial"]
temps = [500, 1500, 2500]
dir = "./results/"
ext = ".txt"
```

```python
def XY(lines, val):
    li = filter(lambda x: int(x[0][1:-1]) == val, lines)
    temp = [(a[1][1:-1], a[2][:-2]) for a in li]
    X = (temp[0][0], temp[len(temp)/2], temp[-1][-1])
    Y = [int(a[-1][1:-2]) for a in li]
    return (X,Y)
```

```python
lines = []
for x in texts:
   lines.append(plotter.split(plotter.lines(dir + x + ext)))
```

```python
di = {}
for i, x in enumerate(lines):
    di[texts[i]] = {}
    for t in temps:
        di[texts[i]][t] = XY(x[1:], t)
```

```python
def barPlotDi(di, title):
    fig, axarr = plt.subplots(3, sharey = True)
    for i, key in enumerate(temps):
        ind = np.arange(len(di[key][1]))
        axarr[i].bar(ind, di[key][1], width, color = 'r')
        axarr[i].set_title(title + ": " + str(key))
        axarr[i].set_xticks((0, len(di[key][1])/2, len(di[key][1])))
        axarr[i].set_xticklabels(di[key][0])
```

```python
for x in texts:
    barPlotDi(di[x], x)
    plt.show()
```
